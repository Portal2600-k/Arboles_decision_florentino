```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor, plot_tree
from sklearn.metrics import mean_absolute_error
```

```python
data = pd.read_csv("C:/Users/Florentino/Documents/dataset/Arboles/Wine.data")
data.head()
```

```python
predictors_col = ["Malic_acid", "Ash", "Acl", "Mg", "Phenols", "Flavanoids", "Nonflavanoid_phenols", "Proanth", "Color_int", "Hue", "OD", "Proline"]
target_col = ['Alcohol']
```

```python
predictors = data[predictors_col]
target = data[target_col]
```

```python
plt.hist(data.Alcohol)
```

```python
X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.25, random_state=42)
```

```python
tree = DecisionTreeRegressor(random_state=42)
```

```python
tree.fit(X_train, y_train)
```

```python
y_pred = tree.predict(X_test)
```

```python
plt.figure(figsize=(15, 10))
plot_tree(tree, feature_names=predictors.columns.to_list(), filled=True)
plt.show()
```

```python
mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error:", mae)
```
